## Chargement des modules et des fonctions de base

```python
# import modules
import numpy as np
from math import *
import matplotlib.pyplot as plt
%matplotlib inline
import pylab as pl
from IPython import display
import time
```

```python
# 2D discrete gradient 
def grad(u):
    ny,nx = u.shape
    Gx = np.zeros(u.shape)
    Gy = np.zeros(u.shape)
    Gx[:,0:nx-1] = u[:,1:nx] - u[:,0:nx-1]
    Gy[0:ny-1,:] = u[1:ny,:] - u[0:ny-1,:]
    return Gx,Gy
```

```python
# divergence of a 2D discrete field
def div(px,py):
    ny,nx = px.shape
    # process the first component (px) of the input field
    div_x = np.zeros(px.shape)
    div_x[:,1:nx-1] = px[:,1:nx-1] - px[:,0:nx-2]
    div_x[:,0] = px[:,0]
    div_x[:,nx-1] = -px[:,nx-2]
    # process the second component (py) of the input field
    div_y = np.zeros(px.shape)
    div_y[1:ny-1,:] = py[1:ny-1,:] - py[0:ny-2,:]
    div_y[0,:] = py[0,:]
    div_y[ny-1,:] = -py[ny-2,:]
    # return output divergence
    return div_x + div_y
```

```python
# Tikhonov based image denoising (see TP 2)
def gaussian_denoise(u0,lbda,niter,alpha,Ndisplay=20,video=False):

    # initialize local variables
    ny,nx = u0.shape
    u = np.copy(u0)
    gx,gy = grad(u)
    E = np.zeros([niter+1,1])
    E[0] = np.sum((u-u0)**2 + lbda*(gx**2+gy**2))
   
    # initialize display (if needed)
    if video:
        plt.figure(figsize=(7,7))
        plt.imshow(u,cmap='gray')
        plt.title("itération 0: E = %.10e" % E[0])
    
    #############
    # main loop #
    #############
    for iter in range(1,niter+1):
        
        # update image u
        g = 2*(u-u0) - 2*lbda*div(gx,gy)
        u = u - alpha*g
        
        # update image gradient
        gx,gy = grad(u)
        
        # compute energy of u
        E[iter] = np.sum((u-u0)**2 + lbda*(gx**2+gy**2))
        
        # update display
        if video and (0 == iter%Ndisplay):
            plt.imshow(u,cmap='gray')
            display.clear_output(wait=True)
            display.display(pl.gcf())
            plt.title("itération %d : E = %.10e" % (iter,E[iter]))
            time.sleep(.1)
    
    # return outputs
    return u,E
```

## Exercice 1 : débruitage d’images par minimisation de la variation totale 

Étant donnée une image bruitée $u_0 : \Omega\to\mathbb{R}$, on s'intéresse au problème
de débruitage

$\DeclareMathOperator*{\argmin}{argmin~}$
\begin{equation}
\overline{u} = \argmin_{u : \Omega \to \mathbb{R}} \underset{= E(u)}{\underbrace{\frac{1}{2}\|u-u_0\|_2^2 + \lambda \mathrm{TV}(u)}}\,,\qquad (\mathcal{P})
\end{equation}

où $\lambda > 0$ et
$$\mathrm{TV}(u) = \displaystyle{\sum_{(x,y) \in \Omega}} \|(\nabla u)
(x,y)\|_2\,.$$


Nous avons démontré en cours la relation
$\overline{u} = u_0 - \lambda \overline{w}$ où $\overline{w}$ désigne la
solution du _problème dual_

\begin{equation}
  \label{eq:map_tv_dual}
  %
  \overline{w} = \argmin_{w \in \mathcal{C}} \frac{1}{2}\|w - (u_0/\lambda)\|_2^2 \,,\qquad (\mathcal{P}^\star)
  %
\end{equation}

où

\begin{equation}
  \label{eq:C}
  %
  \mathcal{C} := \left\{ -\mathrm{div}(p) ~\left/~ p : \Omega\to \mathbb{R}^2 \text{
        et } \max_{(k,\ell)\in\Omega} \|p(k,\ell)\|_2 \leq 1 \right.\right\}\,.
  %
\end{equation}


1-a) Montrer que les problèmes $(\mathcal{P})$ et $(\mathcal{P}^\star)$ admettent chacun une unique solution.

** Problème $(\mathcal{P})$ **

La fonction objectif $u \mapsto \dfrac{1}{2}\|u-u_0\|_2^2 + \lambda \, \mathrm{TV}(u)$ est propre, continue et coercive sur $\mathbb{R}^\Omega$, le problème $(\mathcal{P})$ admet donc au moins une solution. 

Cette fonction objectif est de plus strictement convexe sur $\mathbb{R}^\Omega$, le problème $(\mathcal{P})$ admet donc au plus une solution. 

** Problème $(\mathcal{P}^\star)$ **

**$\bullet$ méthode 1** : l'ensemble $\mathcal{C} \subset \mathbb{R}^\Omega$ est un fermé borné (en tant qu'image par l'application continue $-\mathrm{div}$ du fermé borné $\mathcal{B} = \{p : \Omega \to \mathbb{R}^2 \,,~ \max_{(x,y)\in \Omega} \|p(x,y)\|\leq 1\}$ ), il est donc compact (puisque $\mathbb{R}^\Omega$ est de dimension finie). La fonction $J : p \mapsto \frac{1}{2} \|w - (u_0/\lambda)\|_2^2$ est continue sur le compact $\mathcal{C}$, elle est donc bornée et atteint ses bornes dans $\mathcal{C}$. L'unicité de son minimiseur découle de la stricte convexité de $J$. 

**$\bullet$ méthode 2** : minimiser $p \mapsto \frac{1}{2}\|w - u_0/\lambda\|_2^2$ sur $\mathcal{C}$ revient à minimiser 
$F := p \mapsto \frac{1}{2}\|w - u_0/\lambda\|_2^2 + \delta_\mathcal{C}(p)$ sur $\mathbb{R}^\Omega$. La fonction $F$ est convexe, propre, semi-continue inférieurement et coercive sur $\mathbb{R}^\Omega$, elle admet donc au moins un minimiseur (cours n. 2). Comme $F$ est strictement convexe, elle admet au plus un minimiseur sur $\mathbb{R}^\Omega$.

1-b) On admet pour le moment que l'algorithme du gradient projeté appliqué au
  problème dual $(\mathcal{P}^\star)$ correspond au schéma itératif

$$
    \left\{
      \begin{array}{ccc}
        u^n &= &\lambda \cdot \mathrm{div}(p^n) + u_0 \\[10pt]
        \forall (k,\ell) \in\Omega\,,\quad p^{n+1}(k,\ell) &= &\dfrac{p^n(k,\ell) + \frac{\alpha}{\lambda} (\nabla u^n)(k,\ell)}{\max{\left(1,\|p^n(k,\ell) + \frac{\alpha}{\lambda} (\nabla u^n)(k,\ell)\|_2\right)}}\,,
      \end{array}
    \right.\qquad(\mathcal{S})
$$

Écrire une fonction `tvdenoise(u0,lambda,alpha,niter,...)` permettant d'approcher une solution de $(\mathcal{P})$ en implémentant ce schéma (renvoyer également en sortie la suite de l'énergie des itérés $(E(u^n))_{n \geq 0}$ générés par ce schéma.


```python
def tvdenoise(u0,lbda,niter,alpha,Ndisplay=20,video=False):

    # initialize local variables
    ny,nx = u0.shape
    px = np.zeros([ny,nx])
    py = np.zeros([ny,nx])
    E = np.zeros([niter,1])
   
    # initialize display (if needed)
    if video:
        plt.figure(figsize=(7,7))
        plt.imshow(u0,cmap='gray')
        plt.title("itération 0: E = %.10e" % E[0])
    
    #############
    # main loop #
    #############
    for iter in range(niter):
        
        # update primal variable (u)
        u = lbda*div(px,py) + u0
        
        # update dual variable (p)
        gx,gy = grad(u)
        px = px + (alpha/lbda)*gx
        py = py + (alpha/lbda)*gy
        denom = np.maximum(np.sqrt(px**2+py**2),1)
        px = px / denom
        py = py / denom
                
        # compute energy of u
        E[iter] = np.sum(0.5*(u-u0)**2 + lbda*np.sqrt(gx**2+gy**2))
        
        # update display
        if video and (0 == iter%Ndisplay):
            plt.imshow(u,cmap='gray')
            display.clear_output(wait=True)
            display.display(pl.gcf())
            plt.title("itération %d : E = %.10e" % (iter,E[iter]))
            time.sleep(.1)
    
    # return outputs
    return u,E
```

c) Tester l'algorithme sur l'image `jo.tif` dégradée par un bruit additif blanc gaussien d'écart-type $\sigma = 10$ avec le réglage $\alpha = 0.2$ et $\lambda = 10$. Afficher l'image avant/après débruitage, ainsi que l'évolution de l'énergie au fil des itérations.

```python
# loading reference image
ref = np.double(plt.imread("artifice.tiff"))

# compute noisy observation of the reference image
u0 = ref + np.random.normal(scale=20,size=ref.shape)

# denoise u0 using the tvdenoise algorithm
v,E = tvdenoise(u0,lbda=12,niter=500,alpha=0.2,Ndisplay=10,video=True)
```

```python
# display reference, noisy & denoised images 
fig = plt.figure(figsize=(15,15))

fig.add_subplot(1,3,1)
plt.imshow(u0,cmap='gray') 
plt.title('noisy image')

fig.add_subplot(1,3,2)
plt.title('denoised image')
plt.imshow(v,cmap='gray') 

fig.add_subplot(1,3,3)
plt.title('reference image (ground truth)')
plt.imshow(ref,cmap='gray') 

# Control energy decrease
fig = plt.figure()
plt.plot(E)
plt.xlabel("iteration")
plt.ylabel("energy")
plt.title("energy decrease")
```

d) Comparer cet algorithme de débruitage à celui développé lors du TP n$^\circ$2 (méthode de Tikhonov pour le débruitage gaussien, le code développé en TP2 est rappelé en début de ce TP). 

**NB** : Afin d'effectuer une comparaison équitable de ces deux algorithmes, vous optimiserez pour chaque algorithme le réglage du paramètre $\lambda$ afin de minimiser l'écart quadratique entre l'image restaurée et l'image sans bruit.

```python
# find optimal lambda value for the tvdenoise model
Nsample = 11
lambda_tv = np.linspace(10,20,Nsample)
mse_tv = np.zeros(Nsample)

for id in range(Nsample):
    v,E = tvdenoise(u0,lbda=lambda_tv[id],niter=500,alpha=0.2,video=False)
    mse_tv[id] = np.mean((v-ref)**2)
    print('lambda = ',lambda_tv[id],' : MSE = ',mse_tv[id],sep="")
```

```python
# retrieve optimal lambda for TV denoising 
id = np.where(mse_tv == np.min(mse_tv))
lbda_tv_opt = lambda_tv[id][0]
print('optimal lambda = ',lbda_tv_opt,sep="")

# plot MSE curve 
fig1 = plt.figure()
plt.plot(lambda_tv,mse_tv)
plt.xlabel('lambda')
plt.ylabel('MSE')
plt.title('TV denoise')
```

```python
# find optimal lambda value for the Tikhonov model
Nsample = 11
lambda_tik = np.linspace(0.1,0.5,Nsample)
mse_tik = np.zeros(Nsample)

for id in range(Nsample):
    v,E = gaussian_denoise(u0,lbda=lambda_tik[id],niter=500,alpha=0.1,video=False)
    mse_tik[id] = np.mean((v-ref)**2)
    print('lambda = ',lambda_tik[id],' : MSE = ',mse_tik[id],sep="")
```

```python
# retrieve optimal lambda for Tikhonov model 
id = np.where(mse_tik == np.min(mse_tik))
lbda_tik_opt = lambda_tik[id][0]
print('optimal lambda = ',lbda_tik_opt,sep="")

# plot MSE curve 
fig1 = plt.figure()
plt.plot(lambda_tik,mse_tik)
plt.xlabel('lambda')
plt.ylabel('MSE')
plt.title('Tikhonov denoising')
```

```python
# compute optimal TV denoising result
vopt_tv,E = tvdenoise(u0,lbda=lbda_tv_opt,niter=500,alpha=0.2,video=False)

# compute optimal Tikhonov denoising result
vopt_tik,E = gaussian_denoise(u0,lbda=lbda_tik_opt,niter=500,alpha=0.1,video=False)
```

```python
# display results 
fig = plt.figure(figsize=(20,20))

fig.add_subplot(1,2,1)
plt.title('TV denoising (optimal tuning)')
plt.imshow(vopt_tv,cmap='gray') 

fig.add_subplot(1,2,2)
plt.title('Tikhonov denoising (optimal tuning)')
plt.imshow(vopt_tik,cmap='gray') 
```

## Exercice 2 : construction du schéma dual pour la minimisation de la variation totale


Le but de cet exercice est de démontrer que le schéma numérique $(\mathcal{S})$ proposé en Exercice 1
correspond effectivement à un schéma de gradient projeté appliqué au problème
dual $(\mathcal{P}^\star)$.

2-a) En remarquant que $\mathcal{C} = - \mathrm{div}(\mathcal{B})$ où $\mathcal{B} = \left\{p : \Omega\to \mathbb{R}^2 ~\left/~ \max_{(k,\ell)\in\Omega} \|p(k,\ell)\|_2 \leq 1 \right.  \right\}$, montrer que la solution du problème duale $(\mathcal{P}^\star)$ satisfait

$$
    \overline{w} = -\mathrm{div}(\overline{p}) \quad \text{où} \quad
    \overline{p} = \argmin_{p \in \mathcal{B}} \frac{1}{2} \|\mathrm{div}{(p)} +
    (u_0/\lambda)\|_2^2\, \qquad (\mathcal{P}')
$$


**Réponse** : on a 

$$\overline{w} = \argmin_{w \in -\mathrm{div}(\mathcal{B})} ~\frac{1}{2} \|w - (u_0/\lambda)\|_2^2 = \argmin\limits_{\substack{w = -\mathrm{div}(p) \\ p \in \mathcal{B}}} ~\frac{1}{2} \|-\mathrm{div}(p) - (u_0/\lambda)\|_2^2$$

ainsi 

$$\overline{w} = -\mathrm{div}(\overline{p}) \quad \text{où} \quad \overline{p} \in \argmin_{p \in \mathcal{B}} ~\frac{1}{2} \|\mathrm{div}(p) + (u_0/\lambda)\|_2^2\,.$$

2-b) En notant $J(p) = \frac{1}{2} \|\mathrm{div}{(p)} + u_0/\lambda\|_2^2$ et $\Pi_{\mathcal{B}}$ la projection sur le convexe $\mathcal{B}$, l'algorithme du gradient projeté appliqué au problème $(\mathcal{P}')$ correspond au schéma itératif

$$
p^{n+1} = \Pi_{\mathcal{B}}(p^n - \tau \nabla J(p^n))
$$

Calculer $\nabla J(p^n)$ et montrer que le schéma peut s'écrire

$$
\left\{
    %
    \begin{array}{cl}
      %
      u^n &= \lambda \cdot \mathrm{div}(p^n) + u_0 \\
      p^{n+1} &= \Pi_{\mathcal{B}}\left(p^n + \dfrac{\tau}{\lambda} \cdot \nabla u^n\right)
      %
    \end{array}
    %
\right.
$$


**Réponse** : en utilisant $\mathrm{div} = -\nabla^*$, ou de manière équivalente, $\mathrm{div}^* = -\nabla$, on obtient 

$$\nabla J(p^n) = -\nabla \left(\mathrm{div}(p^n) + u_0/\lambda\right) = -\frac{1}{\lambda} \cdot \nabla \left(\lambda \cdot \mathrm{div}(p^n) + u_0\right) $$

de sorte que le schéma itératif du gradient conjugué s'écrit 

$$p^{n+1} = \Pi_{\mathcal{B}}\left(p^n - \tau \nabla J (p^n)\right) = \Pi_{\mathcal{B}}\left(p^n + \frac{\tau}{\lambda} \cdot \nabla \left( \lambda \cdot \mathrm{div}(p^n) + u_0\right) \right)$$

que l'on peut réécrire 

$$
\left\{
    %
    \begin{array}{cl}
      %
      u^n &= \lambda \cdot \mathrm{div}(p^n) + u_0 \\
      p^{n+1} &= \Pi_{\mathcal{B}}\left(p^n + \dfrac{\tau}{\lambda} \cdot \nabla u^n\right)
      %
    \end{array}
    %
\right.
$$

2-c) Expliciter $\Pi_{\mathcal{B}}$ afin de retrouver le schéma $(\mathcal{S})$

**Réponse** : par définition de la projection sur $\mathcal{B}$, pour tout $p_0 \in \mathbb{R}^{\Omega}\times\mathbb{R}^{\Omega}$, on a 

$$
\Pi_{\mathcal{B}}(p_0) = \argmin_{p \in \mathcal{B}} ~\|p-p_0\|_2^2 = \argmin\limits_{\substack{p \in \mathbb{R}\times \mathbb{R} \\ \forall (k,\ell) \in \Omega \,,~ \|p(k,\ell)\|_2 \leq 1}} ~\sum_{(k,\ell) \in \Omega} \|p(k,\ell)-p_0(k,\ell)\|_2^2\,,
$$

On doit donc minimiser une somme de termes $\|p(k,\ell)-p_0(k,\ell)\|_2^2$ indépendants entre eux, sous-contraintes également indépendantes sur les $p(k,\ell)$, cela revient à minimiser indépendamment chaque terme $\|p(k,\ell)-p_0(k,\ell)\|_2^2$ sous la contrainte $\|p(k,\ell)\|_2 \leq 1$. Ainsi, pour tout $(k,\ell) \in \Omega$, on a 

$$
\Pi_{\mathcal{B}}\left(p_0\right)(k,\ell) = \argmin\limits_{\substack{p(k,\ell) \in \mathbb{R}^2 \\ \|p(k,\ell)\|_2 \leq 1}} ~\|p(k,\ell) - p_0(k,\ell)\|_2^2 = 
\left\{\begin{array}{cl}
p_0(k,\ell) & \text{si } \|p_0(k,\ell)\|_2 \leq 1\\
\dfrac{p_0(k,\ell)}{\|p_0(k,\ell)\|_2}& \text{sinon}
\end{array}\right.
$$

que l'on peut simplifier en 

$$\Pi_\mathcal{B}(p_0)(k,\ell) = \dfrac{p_0(k,\ell)}{\max{\left(1,\|p_0(k,\ell)\|\right)}}$$

Ainsi, en remplaçant $p_0$ par $p^n + \dfrac{\tau}{\lambda} \cdot \nabla u^n$, le schéma précédent

$$
\left\{
    %
    \begin{array}{cl}
      %
      u^n &= \lambda \cdot \mathrm{div}(p^n) + u_0 \\
      p^{n+1} &= \Pi_{\mathcal{B}}\left(p^n + \dfrac{\tau}{\lambda} \cdot \nabla u^n\right)
      %
    \end{array}
    %
\right.
$$

peut se réécrire 

$$
\left\{
\begin{array}{ccc}
u^n &= &\lambda \cdot \mathrm{div}(p^n) + u_0 \\[10pt]
\forall (k,\ell) \in\Omega\,,\quad p^{n+1}(k,\ell) &= &\dfrac{p^n(k,\ell) + \tau/\lambda (\nabla u^n)(k,\ell)}{\max{\left(1,\|p^n(k,\ell) + \tau/\lambda (\nabla u^n)(k,\ell)\|_2\right)}}\,,
\end{array}
\right.
$$
